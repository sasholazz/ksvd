```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Завантаження даних
df = pd.read_csv("/content/shopping_trends.csv")

# Перевірка розміру таблиці
print("Розмір таблиці:", df.shape)

# Перевірка наявності пропусків
print("\nПропуски:")
print(df.isnull().sum())

# Перевірка наявності дублікатів
print("\nДублікати:", df.duplicated().sum())

# Зміна назв стовпців
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Перевірка типів стовпців
print("\nТипи стовпців:")
print(df.dtypes)
```

```python
#1.Гістограма віку (age):
sns.histplot(df['age'], bins=20)
plt.title('Гістограма віку')
plt.xlabel('Вік')
plt.ylabel('Частота')
plt.show()
```

```python
#2.Порівняння розподілу покупок за статтею (item_purchased)
plt.figure(figsize=(10,6))
sns.countplot(y='item_purchased', data=df, order=df['item_purchased'].value_counts().index)
plt.title('Розподіл покупок за статтею')

plt.xlabel('Кількість покупок')
plt.ylabel('Товар')
plt.show()
```

```python
#3. Залежність між віком та сумою покупок (purchase_amount)
plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='purchase_amount_(usd)', data=df)
plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупки (USD)')
plt.show()
```

```python
#4.Графік розподілу оцінок товарів (review_rating)
plt.figure(figsize=(8,5))
sns.histplot(df['review_rating'], bins=5, discrete=True)
plt.title('Графік розподілу оцінок товарів')
plt.xlabel('Оцінка')
plt.ylabel('Частота')
plt.show()
```

```python
#5.Категоріальна діаграма для статі (gender)
plt.figure(figsize=(8,5))
sns.countplot(x='gender', data=df)
plt.title('Категоріальна діаграма для статі')
plt.xlabel('Стать')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
#6.Залежність між рейтингом і сумою покупок (review_rating і purchase_amount)
plt.figure(figsize=(10,6))
sns.scatterplot(x='review_rating', y='purchase_amount_(usd)', data=df)
plt.title('Залежність між рейтингом і сумою покупок')
plt.xlabel('Рейтинг')
plt.ylabel('Сума покупки (USD)')
plt.show()
```

```python
#7.Сезонність покупок (season)
plt.figure(figsize=(8,5))
sns.countplot(x='season', data=df, order=df['season'].value_counts().index)
plt.title('Сезонність покупок')
plt.xlabel('Сезон')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
#8.Графік розподілу розмірів одягу (size)
plt.figure(figsize=(10,6))
sns.countplot(x='size', data=df, order=df['size'].value_counts().index)
plt.title('Графік розподілу розмірів одягу')
plt.xlabel('Розмір')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
#9.Кількість покупок відносно типу оплати (payment_method)
plt.figure(figsize=(10,6))
sns.countplot(x='payment_method', data=df, order=df['payment_method'].value_counts().index)
plt.title('Кількість покупок відносно типу оплати')
plt.xlabel('Метод оплати')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
#10.Залежність між кількістю попередніх покупок (previous_purchases) і сумою поточної покупки (purchase_amount)
plt.figure(figsize=(10,6))
sns.scatterplot(x='previous_purchases', y='purchase_amount_(usd)', data=df)
plt.title('Залежність між кількістю попередніх покупок і сумою поточної покупки')
plt.xlabel('Кількість попередніх покупок')
plt.ylabel('Сума покупки (USD)')
plt.show()
```

```python
#Завдання 2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Генеруємо дані
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Побудова 3D поверхні
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Побудова поверхні
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)  # Додавання кольорової легенди

# Налаштування відображення
ax.set_title('3D Surface Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Показ графіка
plt.show()
```
