```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Bidirectional, LSTM

N = 10000
x = np.linspace(0, 10, N)
data = np.cos(3*x) + 0.1*np.random.randn(N)

model = Sequential()
model.add(Bidirectional(LSTM(128, return_sequences=True), input_shape=(None, 1)))
model.add(Bidirectional(LSTM(128)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

X = data[:-1].reshape(-1, 1, 1)
y = data[1:].reshape(-1, 1)

model.fit(X, y, epochs=20, batch_size=32)

predictions = model.predict(X)

plt.plot(x[1:], y, label='Справжні дані')
plt.plot(x[1:], predictions, label='Прогнозовані дані')
plt.legend()
plt.show()
```
